```python
# Loading the minimum wage
df = pd.read_csv('MinimumWageData.csv', encoding="latin1")
# Filter the df to have  a range between [2010-2019]
df = df[(df['Year'] < 2020) & (df['Year'] > 2009)]
```

```python

nan_columns = df.columns[df.isna().any()].tolist()
print(nan_columns)
print(df[df.isna().any(axis=1)])
print(len(df[df.isna().any(axis=1)]))
```

```python
# Dado que se tiene que footnote es el pie de pagina del sitio web, esto no importa, se dropeara
df.drop('Footnote', axis=1, inplace=True)
```

```python
# Las otras columnas se rellenan con la mediana dado que realmente solo son 4 filas las que tienen nulos en las columnas entonces por ello
# ['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars', 'Department.Of.Labor.Cleaned.High.Value.2020.Dollars']
median_value = df['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'].median()
df['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'] = df['Department.Of.Labor.Cleaned.Low.Value.2020.Dollars'].fillna(median_value)
median_value = df['Department.Of.Labor.Cleaned.High.Value.2020.Dollars'].median()
df['Department.Of.Labor.Cleaned.High.Value.2020.Dollars'] = df['Department.Of.Labor.Cleaned.High.Value.2020.Dollars'].fillna(median_value)
```

```python
# Chequear si hay valores nulos
nan_columns = df.columns[df.isna().any()].tolist()
print(nan_columns)
print(df[df.isna().any(axis=1)])
print(len(df[df.isna().any(axis=1)]))
print(len(df))
```
